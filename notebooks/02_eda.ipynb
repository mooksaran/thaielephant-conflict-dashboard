```python
# 02_eda.ipynb

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# สำหรับกราฟภาษาไทย
plt.rcParams['font.family'] = 'Tahoma'  # หรือ 'Sarabun' / 'Kanit' ถ้ามีในเครื่อง

# โหลดข้อมูลรวมที่ทำไว้ก่อนหน้า
df = pd.read_csv('incidents_combined.csv', parse_dates=['วันที่'])

# แสดงตัวอย่าง
df.head()
```

```python
df['เดือน'] = df['วันที่'].dt.to_period('M')
monthly_counts = df.groupby(['เดือน', 'เหตุการณ์']).size().unstack().fillna(0)

# วาดกราฟ
monthly_counts.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Monthly number of injury/death incidents caused by Thai wild elephants')
plt.xlabel('Month')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
province_counts = df.groupby(['จังหวัด', 'เหตุการณ์']).size().unstack().fillna(0).sort_values('เสียชีวิต', ascending=False)

province_counts.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Number of incidents according to provinces')
plt.ylabel('Number of incidents')
plt.xlabel('Province')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
forest_group = df.groupby(['กลุ่มป่า', 'เหตุการณ์']).size().unstack().fillna(0)

forest_group.plot(kind='barh', stacked=True, figsize=(10, 5))
plt.title('Forest zones with high incident frequency')
plt.xlabel('Number of incidents')
plt.tight_layout()
plt.show()
```

```python
sex_counts = df['เพศ'].value_counts()

plt.pie(sex_counts, labels=sex_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Gender distribution of incident victims')
plt.axis('equal')
plt.show()
```
